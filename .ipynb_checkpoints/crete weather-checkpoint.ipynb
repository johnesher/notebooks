```python
import pandas
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df = pandas.read_csv(
    'LGIR.csv',
    na_values=['M'],
    parse_dates=[0]
)
```

```python
_ = df.sknt.plot()
```

```python
df['year'] = df['valid'].dt.year
df['month'] = df['valid'].dt.month
df['week'] = df['valid'].dt.week
df['day'] = df['valid'].dt.dayofyear

g = df.groupby(['week', 'year'])
weekly_averages = g.aggregate({'sknt': np.mean})
_ = weekly_averages.plot(figsize=(15,10))
#g.tail()
```

```python
#plt.figure(figsize=(20,10))
#import matplotlib
#matplotlib.style.use('ggplot')
g = df.groupby(['month', 'year'])
#g = df.groupby(['day'])
averages = g.aggregate({'gust': np.mean})
_ = averages.plot.bar(figsize=(20,10))
averages = g.aggregate({'sknt': np.mean})
_ = averages.plot.bar(figsize=(20,10))
```
