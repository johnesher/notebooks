```python
import pandas
import matplotlib.pyplot as plt
import numpy as np
# heard that notebook allowed interactive but it just stops show
%matplotlib inline 
```

```python
# lgir is iraklion, lgsa is khania
location = ['iraklion', 'khania'][1]
locations = {'iraklion': 'LGIR.csv', 'khania': 'LGSA.csv'}
df = pandas.read_csv(
    locations[location],
    na_values=['M'],
    parse_dates=[0]
)
```

```python
_ = df.sknt.plot()
```

```python
df['year'] = df['valid'].dt.year
df['month'] = df['valid'].dt.month
df['week'] = df['valid'].dt.week
df['day'] = df['valid'].dt.dayofyear

g = df.groupby(['day', 'year'])
daily_averages = g.aggregate({'sknt': np.mean})
#_ = weekly_averages[189:207].plot.bar()
_ = daily_averages.unstack()[189:207].plot.bar(figsize=(15,10), title=location)
```

```python
#plt.figure(figsize=(20,10))
#import matplotlib
#matplotlib.style.use('ggplot')
g = df.groupby(['month', 'year'])
#g = df.groupby(['day'])
averages = g.aggregate({'gust': np.mean})
_ = averages.unstack().plot.bar(figsize=(20,10), title='gusts at %s kts' % location)
#_ = averages.unstack().boxplot()
averages = g.aggregate({'sknt': np.mean})
usd = averages.unstack()
_ = usd.plot.bar(figsize=(20,10), title='wind speed at %s kts' % location, grid=True)
#_ = usd.boxplot(figsize=(20,10))
```

```python
import seaborn as sns
#ax = sns.boxplot(data = g, x='month',y='year')
#_ = usd.boxplot(figsize=(20,10))
#g.tail()
# g.boxplot produces loadsa boxes and is very slow
ax = df.boxplot(column=['gust'], by=[ 'week'], figsize=(15,10))  #, ylim=(0,40))
#ax[0].set_ylim((0, 40)), ax[1].set_ylim((0, 60))
ax.set_ylim((0, 60))
ax = df.boxplot(column=['sknt'], by=[ 'week'], figsize=(15,10))  #, ylim=(0,40))
#ax[0].set_ylim((0, 40)), ax[1].set_ylim((0, 60))
ax.set_ylim((0, 40))
```

```python
# Try to reduce the dataframe to only the wanted days of the year, 180 to 240 (july & aug)
# 196 to 228 is mid jul to mid aug, C must be back by 15Aug
df_days = df[(df['day']>180) & (df['day']<240)]
ax = df_days.boxplot(column=['gust'], by=['day'], figsize=(15,10), rot=90, fontsize=10)
ax.axvspan(209-180, 225-180, facecolor='g', alpha=0.05)
#ax[0].set_ylim((0, 40)), ax[1].set_ylim((0, 60))
#ax.set_ylim((0, 40))
ax = df_days.boxplot(column=['sknt'], by=['day'], figsize=(15,10), rot=90, fontsize=10)  #, ylim=(0,40))
#ax[0].set_ylim((0, 40)), ax[1].set_ylim((0, 60))
#ax.set_ylim((0, 40))
ax.axvspan(209-180, 225-180, facecolor='g', alpha=0.05)
```
