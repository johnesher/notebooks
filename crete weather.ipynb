```python
import pandas
import matplotlib.pyplot as plt
import numpy as np
# heard that notebook allowed interactive but it just stops show
%matplotlib inline 
```

```python
df = pandas.read_csv(
    'LGSA.csv',
    na_values=['M'],
    parse_dates=[0]
)
```

```python
_ = df.sknt.plot()
```

```python
df['year'] = df['valid'].dt.year
df['month'] = df['valid'].dt.month
df['week'] = df['valid'].dt.week
df['day'] = df['valid'].dt.dayofyear

g = df.groupby(['day', 'year'])
weekly_averages = g.aggregate({'sknt': np.mean})
#_ = weekly_averages.plot(figsize=(15,10))
_ = weekly_averages.unstack()[189:207].plot.bar(figsize=(15,10))
```
