```python
with open(target) as infile:
    Xs, Ys = [], []
    for li_no, line in enumerate(infile.readlines()):
        if 'SUCCESS' in line:
            Xs += [li_no]
            Ys += [1]
        if 'batty' in line:
            Xs += [li_no]
            Ys += [2]
        if 'notifying' in line:
            Xs += [li_no]
            Ys += [3]
        if 'asking' in line:
            Xs += [li_no]
            Ys += [4]
plt.plot(Xs, Ys, 's', marker='.')
```
