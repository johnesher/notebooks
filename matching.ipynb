Estimated values for X by shifting X values by -20 so the change from cap to ind happens at the resonance rather than at lower frequency.

At 12M impedance is about 1.8-22j (after offset 2 -2.7j)

At 20M impedance is about 11+31j (after offset 2 +6.5j)

$C = \frac{1}{\omega X_C}$

$L = \frac{X_L}{\omega}$

```python
wc = 2 * 3.1415 * 10e6  # 12e6
C_offset = 1/(wc*2.7)  # 1/(wc*22)
wl = 2 * 3.1415 * 20e6
L_offset = 6.5 / wc  # 31 / wc
w_resonance = (1/(C_offset * L_offset))**0.5
f_resonance = w_resonance / (2 * 3.14159)
print('%.2fnH, %.2fpF, %.2fMHz' % (L_offset * 1e9, C_offset * 1e12, f_resonance / 1e6))
```

The impedance is really that of the L and C simultaneously. Presume that the circuit is a parallel LC fed by a series resistor. Ignore the resistor and work only with the admittances.

$Y_{LC} = Y_L + Y_C$

We have a different Y at 12M and at 20M. Use these to solve for L and C separately.

$Y_L = \frac{1}{\omega L}$  and  $Y_C = {\omega C}$

$Y_{LC_{10}} = \frac{1}{\omega_{10} L} + {\omega_{10} C}$ 

$Y_{LC_{20}} = \frac{1}{\omega_{20} L} + {\omega_{20} C}$

$C = \frac{Y_{LC_{10}} - \frac{1}{\omega_{10} L}}{\omega_{10}}$

$C = \frac{Y_{LC_{10}}}  {\omega_{10}} - \frac{1}{{\omega_{10}}^2 L}$

Now substitute into the 20MHz equation

$Y_{LC_{20}} = \frac{1}{\omega_{20} L} + \omega_{20}\left( \frac{Y_{LC_{10}}}  {\omega_{10}} - \frac{1}{{\omega_{10}}^2 L} \right) $

$Y_{LC_{20}} = \frac{1}{\omega_{20} L} + \omega_{20}\frac{Y_{LC_{10}}}{\omega_{10}} - \frac{\omega_{20}}{{\omega_{10}}^2 L}$

$Y_{LC_{20}} - \omega_{20}\frac{Y_{LC_{10}}}{\omega_{10}} = \frac{1}{\omega_{20} L} - \frac{\omega_{20}}{{\omega_{10}}^2 L}$

$L \left(Y_{LC_{20}} - \omega_{20}\frac{Y_{LC_{10}}}{\omega_{10}}\right) = \frac{1}{\omega_{20}} - \frac{\omega_{20}}{{\omega_{10}}^2}$

$L = \frac{\frac{1}{\omega_{20}} - \frac{\omega_{20}}{{\omega_{10}}^2}}
{Y_{LC_{20}} - \omega_{20}\frac{Y_{LC_{10}}}{\omega_{10}}}$

Divide through by $\omega_{20}$

$L = \frac
{\frac{1}{{\omega_{20}}^2} - \frac{1}{{\omega_{10}}^2}}
{\frac{Y_{LC_{20}}}{\omega_{20}} - \frac{Y_{LC_{10}}}{\omega_{10}}}$

```python
Y20 = 1 / 6.6
Y10 = 1 / -27
w20 =  2 * np.pi * 20e6
w10 = 2 * np.pi * 10e6
denom_L = (1 / w20) - (w20 / w10**2)
num_L = Y20 - w20 * Y10 / w10
L_offset = denom_L / num_L
C_offset = (Y10 / w10) - 1 / (w10**2 * L_offset)
w_resonance = (1/(C_offset * L_offset))**0.5
f_resonance = w_resonance / (2 * 3.14159)
f_resonance = 0
print('%.2fnH, %.2fpF, %.2fMHz' % (L_offset * 1e9, C_offset * 1e12, f_resonance / 1e6))
Ld = 1/w20**2 - 1/w10**2
Ln = Y20/w20 - Y10/w10
L = Ld/Ln
Y10/w10
Y20/w10
```

Need to be offset phase in phase column so need to convert the rl and phase into s or Z.

$S_{11}$ is simply $\Gamma$, with magnitude $\rho$, and phase $\theta$

The sparam display routine expects s to be in cartesian coords so need to convert the mag and phase to a complex xy.

```python
df['rho'] = 10**(df['Returnloss (dB)']/20)  # mag of gamma
df['phase (rads)'] = np.pi * df['Returnphase (°)'] / 180
df['s11'] = df['rho'] * np.exp(1j * df['phase (rads)'])
df['phase_offsets'] = df['Frequency (Hz)'] * 2 * np.pi * time_offset_15e6
df['s11_offset'] = df['rho'] * np.exp(1j * (df['phase (rads)'] + df['phase_offsets']))
plt.figure(figsize=(9, 9))
ax = plt.subplot(1, 1, 1, projection='smith')
plt.plot(df['s11'], datatype=SmithAxes.S_PARAMETER);
plt.plot(df['s11_offset'], datatype=SmithAxes.S_PARAMETER);
```
